```python
#this code shows how to do time based querying of an api
historical_time = []
counter = 0
time_to_sleep = 1 #definitely need a longer time interval for this api
while(counter < 20):
    response = req.get('http://api.weatherapi.com/v1/current.json', params=params)
    historical_time.append(response.json())
    sleep(time_to_sleep)
    counter += 1
```

```python
historical_time
```

```python
df = pd.json_normalize(historical_time)
df.head()
```

```python
#this code shows how to iterate over many different locations
locations_to_search = ['London', 'Berlin', 'Denver']
location_weather = []
for location in locations_to_search:
    params = {'key':api_key, 'q':location}
    response = req.get('http://api.weatherapi.com/v1/current.json', params=params)
    location_weather.append(response.json())
```

```python
location_weather
```

```python
df = pd.json_normalize(location_weather)
df.head()
```
